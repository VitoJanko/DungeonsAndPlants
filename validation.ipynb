```python
for habitat in ["Arid", "Arctic", "Forest", "Coast", "Mountain", "Cave", "Plains", "Tropics", "Urban"]:
    suma = sum(plants_all[habitat])
    print(f"{habitat} ({suma})")
    
    for i in range(0,7):
        if i!=1:
            possible = sum((plants_all["Profficiency"]<=i) & (plants_all[habitat]))
            count = sum((plants_all["Profficiency"]==i) & (plants_all[habitat]))

            ratio = int((count/possible)*100)
            print(f"Profficiency {i}: {count} ({ratio}%)")
    print("--")
```

```python
result = ""
for proficiency in [0, 2, 3, 4, 5, 6]:
    result += str(proficiency) + " - "
    for i, plant in plants_all[(plants_all["Mountain"]) & (plants_all["Profficiency"] == proficiency)].iterrows():
        result += plant["Name"] + ", "
    result += "\n"
print(result)
```

```python
def format_text(cond):
    cond = cond.strip()
    if len(cond)==0:
        return cond
    cond = list(cond)
    if cond[-1]==".":
        cond[-1]=":"
    else:
        cond +=[":"]
    cond = "".join(cond)
    cond =cond.replace(".",",")
    return cond
```

```python
plants_all["t2a"] = plants_all["t2a"].apply(format_text)
plants_all["t1a"] = plants_all["t1a"].apply(format_text)
```
