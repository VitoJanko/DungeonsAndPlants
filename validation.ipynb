```python
import pandas as pd
```

```python
plants_all = pd.read_csv("plant_list.csv", sep=";").fillna("")
```

```python
plants_all.dtypes
```

```python
def format_text(cond):
    cond = cond.strip()
    if len(cond)==0:
        return cond
    cond = list(cond)
    if cond[-1]==".":
        cond[-1]=":"
    else:
        cond +=[":"]
    cond = "".join(cond)
    cond =cond.replace(".",",")
    return cond
```

```python
plants_all["t2a"] = plants_all["t2a"].apply(format_text)
plants_all["t1a"] = plants_all["t1a"].apply(format_text)
```

```python
plants_all.to_csv("plant_list_mod.csv", sep=";")
```
